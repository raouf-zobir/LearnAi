```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize']  = (10,6)
```

```python
df = pd.read_csv('bhp.csv')
df.head()
```

```python
df.shape
```

```python
# step 1
lower_limit, upper_limit = df.price_per_sqft.quantile([0.001,0.999])
lower_limit, upper_limit
```

```python
df[df.price_per_sqft < lower_limit]
```

```python
df[df.price_per_sqft > upper_limit]
```

```python
df[(df.price_per_sqft > upper_limit) | (df.price_per_sqft < lower_limit)]
```

```python

outliers = df[(df.price_per_sqft < upper_limit) & (df.price_per_sqft > lower_limit)]
outliers.shape
```

```python
df.shape
```

```python
df.shape[0] - outliers.shape[0]
```

```python
# step 2
df2 = df[(df.price_per_sqft < upper_limit) & (df.price_per_sqft > lower_limit)]
df2.shape
```

```python
max_limit = df2.price_per_sqft.mean() + 4 * df2.price_per_sqft.std()
min_limit = df2.price_per_sqft.mean() - 4 * df2.price_per_sqft.std()
max_limit,min_limit
```

```python
df2[df2.price_per_sqft > max_limit]
```

```python
df2[df2.price_per_sqft < min_limit]
```

```python
df2[(df2.price_per_sqft > max_limit) | (df2.price_per_sqft < min_limit)]
```

```python
df3 = df2[(df2.price_per_sqft < max_limit) & (df2.price_per_sqft > min_limit)]
df3.shape
```

```python
df2.shape
```

```python
df2.shape[0] - df3.shape[0]
```

```python
plt.hist(df3.price_per_sqft, bins=20, rwidth=0.8 , density=True)
plt.xlabel('price per sqft (inches)', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.show()
```

```python
from scipy.stats import norm
plt.hist(df3.price_per_sqft, bins=20, rwidth=0.8 , density=True)
plt.xlabel('price per sqft (inches)', fontsize=20)
plt.ylabel('Count', fontsize=20)
rng = np.arange(df3.price_per_sqft.min()-5000,df3.price_per_sqft.max(),100)
plt.plot(rng, norm.pdf(rng,df3.price_per_sqft.mean(),df3.price_per_sqft.std()))
plt.show()
```

```python
df2['zscore'] = (df2.price_per_sqft - df2.price_per_sqft.mean()) / df2.price_per_sqft.std()
df2.shape
```

```python
df2[df2.zscore< -4]
```

```python
df2[df2.zscore > 4]
```

```python
df2[(df2.zscore> 4) | (df2.zscore < -4)]
```

```python
outliers_zscore = df2[(df2['zscore'] < 4) & (df2['zscore'] > -4)]
outliers_zscore.shape
```

```python
df2.shape[0]- outliers_zscore.shape[0]
```
