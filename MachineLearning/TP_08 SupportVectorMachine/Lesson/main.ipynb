```python
# Example
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Filter for two classes (binary classification)
df = df[df['target'] != 2]  # Use only classes 0 and 1
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]

# Features and labels
X = df[["petal length (cm)", "petal width (cm)"]].values
y = df["target"].values

# Train an SVC model
svc = SVC(kernel='linear', C=1)
svc.fit(X, y)

# Create a mesh grid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Compute decision function values
Z = svc.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the scatter plot
plt.figure(figsize=(8, 6))
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0["petal length (cm)"], df0["petal width (cm)"], color='green', marker="+", label="Class 0")
plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"], color='blue', marker="*", label="Class 1")

# Plot the decision boundary and margins
plt.contour(xx, yy, Z, levels=[-1, 0, 1], linestyles=['--', '-', '--'], colors='black')
plt.legend()
plt.title('SVC Decision Boundary')
plt.show()
```
