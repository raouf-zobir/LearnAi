<h2 style='color:yellow' align='center'>L1 and L2 Regularization</h2>

```python
import pandas as pd
```

```python
dataset = pd.read_csv('Housing.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance',
                'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True).astype(float)
dataset.head()
```

```python
X = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=2)
```

<h4 style='color:white' align='center'>Linear Regression</h4>

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train,y_train)
```

```python
linear_reg.score(X_test,y_test)
```

```python
linear_reg.score(X_train,y_train)
```

<h4 style='color:white' align='center'>L1 Lasso Regression</h4>

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

<h4 style='color:white' align='center'>L2 Ridge Regression</h4>

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
